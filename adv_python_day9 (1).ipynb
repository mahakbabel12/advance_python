# ENCODING

```python
#Encoding ---->transform categorical data into Numerical data.

#(1.) Label Encoding ---->apply on single column and here we cannot 
#decide for which sub category what will be the number.

#(2.) Ordinal Encoding --->apply on multiple columns.and here decide the subcategory value.

#(3.)OneHot Encoding ----->it will create new columns based on the sub-categories.

#(4.)Get_dummies(pandas)--->It will create new columns based on the sub-categories.
```

```python
df=pd.read_csv("covid_toy.csv")
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
```

```python
df.head(2)
```

```python
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.sample(5)
```

# get_dummies

```python
a= pd.get_dummies(df , columns =  ['gender' , 'cough' , 'city' , 'has_covid'],drop_first = True)
```

```python
a = a.dropna()
a.head(2)
```

```python
a=a.astype(int)
a
```
